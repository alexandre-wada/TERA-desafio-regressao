# Objetivo 1

```python
df['PrecoVenda_log']=np.log(df['PrecoVenda'])
sns.kdeplot(df['PrecoVenda_log'])
```

```python
list(df.columns)
```

```python
import statsmodels.formula.api as smf

df_reg = df[df['CondicaoVenda']=='Normal']

smf.ols('PrecoVenda_log ~ TamanhoLote + Qualidade + Banheiro + TotalQuartos + ReformaHa + CarrosGaragem', data=df).fit().summary()
```

```python
#Melhor R2, mas tem multicolinearidade
df_reg = df[df['CondicaoVenda']=='Normal']

smf.ols('PrecoVenda_log ~ QuartosPorArea + Qualidade + Bairro + TotalQuartos + ReformaHa + CarrosGaragem', data=df_reg).fit().summary()
```

```python
df_reg = df[df['CondicaoVenda']=='Normal']

smf.ols('PrecoVenda_log ~ TamanhoLote + QualiECondi + Bairro + TotalQuartos + ReformaHa + ClasseZona', data=df).fit().summary()
```

```python
df_reg = df[df['CondicaoVenda']=='Normal']

smf.ols('PrecoVenda_log ~ TamanhoLote ', data=df).fit().summary()
```

```python
#padronizando os dados numéricos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standard = pd.DataFrame(scaler.fit_transform(df[['TotalQuartos', 'ReformaHa', 'QuartosPorArea', 'CarrosGaragem', 'Qualidade' ]]), columns = ['TotalQuartos1', 'ReformaHa1', 'QuartosPorArea1', 'CarrosGaragem1', 'Qualidade1' ] )
df_standard
```

```python
data1 = df.merge(df_standard, left_index=True, right_index=True)

list(data1.columns)
```

```python
df_reg = data1[data1['CondicaoVenda']=='Normal']

smf.ols('PrecoVenda_log ~ QuartosPorArea1 + Qualidade1 + Bairro + TotalQuartos1 + ReformaHa1 + CarrosGaragem1', data=df_reg).fit().summary()
```

Esse ultimo teve o melhor resultado de R2 sem dar o problema da colinearidade.
É possivel ver a influencia positiva da qualidade do imovel e da quantidade total de quartos
Variaveis com influencia negativa são a quantidade de quartos por área do terreno e a quantidade de anos desde a ultima reforma.
Cada bairro tem uma influência diferente, e alguns são bons ou ruins em graus diferentes.
